```python
import gradio as gr
import requests
import json

def Sentiment_Analysis(text):
    url = "https://text-analysis12.p.rapidapi.com/sentiment-analysis/api/v1.1"

    payload = {
      "language": "english",
      "text": text
    }
    headers = {
      "content-type": "application/json",
      "X-RapidAPI-Key": "a7833d2622mshd4e0dce16cf7abcp138a03jsnba1cb6751b51",
      "X-RapidAPI-Host": "text-analysis12.p.rapidapi.com"
    }

    response = requests.request("POST", url, json=payload, headers=headers)
    return_text = json.loads(response.text)

    return return_text['sentiment']

with gr.Blocks() as demo:
    text = gr.Textbox(label="Text")
    sen = gr.Textbox(label="Sentiment Analysis: ")
    greet_btn = gr.Button("Detect sentiment")
    greet_btn.click(fn=Sentiment_Analysis, inputs=text, outputs=sen)

demo.launch()
```
