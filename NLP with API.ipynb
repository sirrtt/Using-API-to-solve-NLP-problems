```python
import gradio as gr
import json
import requests

def Language_Detection(text):

	url = "https://text-analysis12.p.rapidapi.com/language-detection/api/v1.1"

	payload = {"text": text}
	headers = {
		"content-type": "application/json",
		"X-RapidAPI-Key": "1a0e5f9885msh35a0da731bf7501p1e985ajsn84f0decc23c7",
		"X-RapidAPI-Host": "text-analysis12.p.rapidapi.com"
	}

	response = requests.request("POST", url, json=payload, headers=headers)
	data = json.loads(response.text)
	return(data['language_probability'])
 
gr.Interface(fn = Language_Detection, inputs = "text", outputs = "text").launch()
```
